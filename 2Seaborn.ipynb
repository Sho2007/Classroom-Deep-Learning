```python
import seaborn as sns
import matplotlib.pyplot as plt

# สร้างข้อมูล list ตัวเลข
numbers = [12, 15, 14, 18, 19, 20, 21, 18, 16, 14, 13, 15, 17, 18, 22]

categories = ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A']
```

```python
sns.histplot(numbers, kde=False)  # ใช้ KDE เพื่อเพิ่มเส้นความหนาแน่น
plt.show()
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
```

```python
sns.boxplot(x=categories, y=numbers)  # แสดงการกระจายตามหมวดหมู่
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Box Plot")
plt.show()
```

```python
sns.violinplot(x=categories, y=numbers)  # คล้าย Box Plot แต่แสดง KDE ด้วย
plt.xlabel("Category")
plt.ylabel("Value")
plt.title("Violin Plot")
plt.show()
```

```python
time = list(range(len(numbers)))  # ตัวเลขลำดับเวลา

sns.scatterplot(x=time, y=numbers, hue=categories)  # Scatter Plot พร้อม Hue ตามหมวดหมู่
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Scatter Plot")
plt.show()
```

```python
print(list(range(len(numbers))))
```

```python
sns.lineplot(x=time, y=numbers)  # เส้นแสดงแนวโน้ม
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Line Plot")
plt.show()
```

```python
sns.kdeplot(numbers, shade=True)  # การแจกแจงความหนาแน่นแบบต่อเนื่อง
plt.xlabel("Value")
plt.title("KDE Plot")
plt.show()
```

```python
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
sns.set_style("dark")
sns.set_palette("pastel")
sns.set_context("talk")
plt.figure(figsize=(8, 6))
plt.grid(True, linestyle='--', color='gray', alpha=0.7)
plt.rcParams["font.family"] = "Tahoma"
plt.legend(title="Group", loc="upper left")
```
